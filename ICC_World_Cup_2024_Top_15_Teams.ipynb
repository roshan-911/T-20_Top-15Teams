# Introduction

In this notebook, I am going to analyze the top 15 teams with the best win percentages in the ICC World Cup 2024. I am loading the data, cleaning the data, performing exploratory data analysis (EDA), and visualizing key insights.

# Loading the Data

```python
# Import necessary libraries
import pandas as pd

# Load the dataset
file_path = 'Team - Best Win Percentage.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
df.head()
```

# Data Cleaning

```python
# Check for missing values
print("Missing values in each column:")
print(df.isnull().sum())

# Check data types of each column
print("\nData types of each column:")
print(df.dtypes)

# Display summary statistics
print("\nSummary statistics:")
print(df.describe())
```

# Exploratory Data Analysis (EDA)

```python
# Basic information about the dataset
print("\nBasic information about the dataset:")
print(df.info())

# Number of unique teams
unique_teams = df['Team'].nunique()
print(f"\nNumber of nique teams: {unique_teams}")

# Top teams by win percentage
top_teams = df.sort_values(by='Win %', ascending=False)
print("\nTop teams by win percentage:")
print(top_teams)
```

# Visualizations

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# Bar plot of win percentages
plt.figure(figsize=(12, 8))
sns.barplot(x='Win %', y='Team', data=top_teams, palette='viridis')
plt.title('Top 15 Teams by Win Percentage in ICC World Cup 2024')
plt.xlabel('Win Percentage')
plt.ylabel('Team')
plt.show()

# Box plot to show distribution of win percentages
plt.figure(figsize=(12, 8))
sns.boxplot(y='Win %', data=df, palette='viridis')
plt.title('Distribution of Win Percentages among Top 15 Teams')
plt.ylabel('Win Percentage')
plt.show()
```

# Conclusions

In this notebook, we analyzed the top 15 teams with the best win percentages in the ICC World Cup 2024. We loaded and cleaned the data, performed exploratory data analysis, and visualized key insights. 

The bar plot and box plot provided a clear view of the distribution of win percentages among these teams. This analysis helps us understand which teams performed exceptionally well in the tournament.
